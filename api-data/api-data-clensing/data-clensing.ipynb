```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Resourses/newsKeyWords.csv")
df
```

```python
# Adjusting the approach based on the data structure

# Step 1: Identify rows to drop (all columns except the first two are NaN)
# Adjusting to include the 'Company' column in the check as well, so we start checking from the fourth column
mask_drop = df.iloc[:, 3:].isna().all(axis=1)

# Step 2: Drop these rows
df_filtered = df.drop(df[mask_drop].index)

# Step 3: Fill in 'Null' for partially empty rows in remaining DataFrame
# Adjusting to fill NaN values from the fourth column onwards
df_filtered.iloc[:, 3:] = df_filtered.iloc[:, 3:].fillna('Null')

# Display the first few rows of the adjusted DataFrame to verify changes
df_filtered.head(10)


```

```python
df_filtered.to_csv("Resourses/newsKeyWords.csv", index=False)
```
